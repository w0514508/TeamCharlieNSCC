# Weather Station List
This lab is cleaning file for Data Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/raw/Yashaswi/climate_station_list.csv")
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.shape
df.columns
```

```python
df = df.drop(
    columns=[
        'HLY First Year',
        'HLY Last Year',
        'DLY First Year',
        'DLY Last Year',
        'MLY First Year',
        'MLY Last Year',
        'WMO Identifier',
        'TC Identifier',
        'Elevation',
    ]
)
```

```python
df.head(3)
df['Province'].value_counts()
```

```python

atlantic_provinces = [
    'NEWFOUNDLAND AND LABRADOR',
    'NOVA SCOTIA',
    'NEW BRUNSWICK',
    'PRINCE EDWARD ISLAND'
]

df_atlantic = df[df['Province'].isin(atlantic_provinces)]
```

```python
df_atlantic.head()
```

```python
df_atlantic.shape
```

```python
df_atlantic = df_atlantic[df_atlantic['Last Year'].isin([2025, 2026])]
```

```python
df_atlantic.shape
```

```python
df_atlantic.head()
```

```python
df_atlantic['First Year'].drop_duplicates().sort_values()
```

```python
df_atlantic = df_atlantic[df_atlantic['First Year'] > 1992]
```

```python
df_atlantic.shape
```

```python
df_atlantic.info()
```

```python
df_atlantic.head(10)
```

```python
df_atlantic.to_csv("Stationlist_clean.csv", index=False)
```
